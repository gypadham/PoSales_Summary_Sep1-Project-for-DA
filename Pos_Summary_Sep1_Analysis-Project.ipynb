```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the uploaded Excel file to inspect its structure
file_path = '/mnt/data/Pos_Summary_Sep1.xlsx'

# Load the dataset (Excel sheet name 1, "Sale & Collection")

data = pd.read_excel(r'C:\Users\gypad\OneDrive\Desktop\Project related-Testing\Pos_Summary_Sep1.xlsx',sheet_name=1)
# Ensure Invoice Date is in datetime format
data['Invoice date'] = pd.to_datetime(data['Invoice date'], errors='coerce')


# Inspect the unique values in the "Invoice date" column
print(data['Invoice date'].head())  # Display a sample of the column

# Convert "Invoice date" to a datetime format, handling errors
#data['Invoice date'] = pd.to_datetime(data['Invoice date'], errors='coerce')

# Check for rows with invalid dates
invalid_dates = data[data['Invoice date'].isna()]
print("Rows with invalid dates:")
print(invalid_dates)

# Optionally, drop rows with invalid dates if they are not needed
data = data.dropna(subset=['Invoice date'])

# Proceed with the analysis


# Descriptive Analysis
# Summarize sales by branch, brand, or category
sales_by_branch = data.groupby('Branch Name')['Bill Amount'].sum()
sales_by_brand = data.groupby('Brand')['Bill Amount'].sum()
sales_by_category = data.groupby('Category')['Bill Amount'].sum()

# Calculate total sales, average margin, and discounts
summary_stats = {
    'Total Sales': data['Bill Amount'].sum(),
    'Average Margin': data['Mar%'].mean(),
    'Total Discounts': data['Disc%'].sum()
}

# Trend Analysis
# Sales trends over time
sales_trends = data.groupby('Invoice date')['Bill Amount'].sum()

# Profitability Analysis
# Overall and category-specific margins
margin_by_category = data.groupby('Category')['Net Margin Without Gst'].sum()
margin_percentage_by_category = data.groupby('Category')['Mar%'].mean()

# Payment Method Analysis
# Distribution of payment methods
payment_columns = ['Cash', 'HDFEDC', 'PINELABEDC', 'AMEXEDC', 'Benowedc', 'Paytmedc', 'Phonepe', 'Gpay']
payment_distribution = data[payment_columns].apply(pd.to_numeric, errors='coerce').sum()

# Product Performance
# Top-performing products
sales_by_product = data.groupby('Item Name')['Bill Amount'].sum().sort_values(ascending=False)

# Discounts vs Sales
discount_sales_correlation = data[['Disc%', 'Total']].apply(pd.to_numeric, errors='coerce').corr()

# Branch/Employee Analysis
sales_by_employee = data.groupby('Salesperson Name')['Bill Amount'].sum()
sales_by_branch_employee = data.groupby(['Branch Name', 'Salesperson Name'])['Bill Amount'].sum()

# Correlation Analysis
correlation_metrics = data[['Selling Rate Without Gst', 'Net Margin Without Gst', 'Mar%', 'Disc%']].corr()

# Visualizations
plt.figure(figsize=(12, 6))
sales_trends.plot(title='Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x=sales_by_category.index, y=sales_by_category.values)
plt.title('Sales by Category')
plt.xticks(rotation=45)
plt.ylabel('Total Sales')
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x=sales_by_branch.index, y=sales_by_branch.values)
plt.title('Sales by Branch')
plt.xticks(rotation=45)
plt.ylabel('Total Sales')
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x=payment_distribution.index, y=payment_distribution.values)
plt.title('Payment Method Distribution')
plt.xticks(rotation=45)
plt.ylabel('Total Payment')
plt.show()

# Print Summary Statistics
print("Descriptive Analysis")
print(summary_stats)
print("\nProfitability Analysis")
print(margin_by_category)
print("\nCorrelation Analysis")
print(correlation_metrics)
print("\nTop Products")
print(sales_by_product.head(5))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

# Load the dataset
file_path = '/mnt/data/POS_Summary_Sep1.xlsx'
# Load the dataset (Excel sheet name 2, "Margin Sheet")

data = pd.read_excel(r'C:\Users\gypad\OneDrive\Desktop\Project\Pos_Summary_Sep1.xlsx', sheet_name=2)

# Data Cleaning and Preparation
data = data.dropna(how='all')  # Drop completely empty rows
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'], errors='coerce')
data['net_amt'] = pd.to_numeric(data['net_amt'], errors='coerce')
data['Qty'] = pd.to_numeric(data['Qty'], errors='coerce')
data['Net Margin'] = pd.to_numeric(data['Net Margin'], errors='coerce')
data['Disc %'] = pd.to_numeric(data['Disc %'], errors='coerce')

# Aggregate Data for Analysis
sales_by_branch = data.groupby('Branch')['net_amt'].sum().reset_index()
sales_by_brand = data.groupby('Brand')['net_amt'].sum().reset_index()
daily_sales = data.groupby('Invoice Date')['net_amt'].sum().reset_index()
discount_impact = data.groupby('Disc %')['net_amt'].sum().reset_index()

# Dashboard Setup
dashboard = Dash(__name__)

# Layout
dashboard.layout = html.Div([
    html.H1("POS Margin Summary Dashboard", style={'textAlign': 'center'}),

    dcc.Tabs([
        dcc.Tab(label='Sales by Branch', children=[
            dcc.Graph(
                id='sales-branch-chart',
                figure=px.bar(sales_by_branch, x='Branch', y='net_amt', title='Total Sales by Branch')
            )
        ]),

        dcc.Tab(label='Sales by Brand', children=[
            dcc.Graph(
                id='sales-brand-chart',
                figure=px.bar(sales_by_brand, x='Brand', y='net_amt', title='Total Sales by Brand')
            )
        ]),

        dcc.Tab(label='Daily Sales', children=[
            dcc.Graph(
                id='daily-sales-chart',
                figure=px.line(daily_sales, x='Invoice Date', y='net_amt', title='Daily Sales Trend')
            )
        ]),

        dcc.Tab(label='Discount Impact', children=[
            dcc.Graph(
                id='discount-impact-chart',
                figure=px.scatter(discount_impact, x='Disc %', y='net_amt', title='Impact of Discounts on Sales')
            )
        ])
    ])
])

# Run Dashboard
if __name__ == '__main__':
    dashboard.run_server(debug=True, port=8050)
```

```python
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html

# Load the dataset
file_path = '/mnt/data/POS_Summary_Sep1.xlsx'
# Load the dataset (Excel sheet name 0, "Sale Sheet")
data = pd.read_excel(r'C:\Users\gypad\OneDrive\Desktop\Project\Pos_Summary_Sep1.xlsx', sheet_name=0)

# Data Cleaning and Preparation
data = data.dropna(how='all')  # Drop completely empty rows
data['Voucher Date'] = pd.to_datetime(data['Voucher Date'], errors='coerce')
data['Voucher Amount'] = pd.to_numeric(data['Voucher Amount'], errors='coerce')
data['Total Qty'] = pd.to_numeric(data['Total Qty'], errors='coerce')
data['Total Gst Value'] = pd.to_numeric(data['Total Gst Value'], errors='coerce')
data['Cash Amount'] = pd.to_numeric(data['Cash Amount'], errors='coerce')
data['Credit Card Amount'] = pd.to_numeric(data['Credit Card Amount'], errors='coerce')
data['Paytm Amount'] = pd.to_numeric(data['Paytm Amount'], errors='coerce')

# Aggregate Data for Analysis
branch_performance = data.groupby('Branch Name').agg({
    'Voucher Amount': 'sum',
    'Total Qty': 'sum',
    'Total Gst Value': 'sum'
}).reset_index()

customer_performance = data.groupby('Customer Name').agg({
    'Voucher Amount': 'sum',
    'Total Qty': 'sum'
}).reset_index()

salesperson_performance = data.groupby('Salesperson Name').agg({
    'Voucher Amount': 'sum',
    'Total Qty': 'sum'
}).reset_index()

payment_modes = data[['Cash Amount', 'Credit Card Amount', 'Paytm Amount', 'Phonepay Amount', 'Bank Amount']].sum().reset_index()
payment_modes.columns = ['Payment Mode', 'Total Amount']

sales_trend = data.groupby('Voucher Date').agg({
    'Voucher Amount': 'sum',
    'Total Qty': 'sum'
}).reset_index()

cancellation_analysis = data[data['iscancelled'] == 1]

# Dashboard Setup
dashboard = Dash(__name__)

# Layout
dashboard.layout = html.Div([
    html.H1("POS Sales Summary Dashboard", style={'textAlign': 'center'}),

    dcc.Tabs([
        dcc.Tab(label='Branch Performance', children=[
            dcc.Graph(
                id='branch-performance-chart',
                figure=px.bar(branch_performance, x='Branch Name', y='Voucher Amount', title='Sales by Branch')
            )
        ]),

        dcc.Tab(label='Customer Analysis', children=[
            dcc.Graph(
                id='customer-performance-chart',
                figure=px.bar(customer_performance, x='Customer Name', y='Voucher Amount',
                              title='Top Customers by Sales')
            )
        ]),

        dcc.Tab(label='Salesperson Analysis', children=[
            dcc.Graph(
                id='salesperson-performance-chart',
                figure=px.bar(salesperson_performance, x='Salesperson Name', y='Voucher Amount',
                              title='Sales Contribution by Salesperson')
            )
        ]),

        dcc.Tab(label='Payment Modes', children=[
            dcc.Graph(
                id='payment-modes-chart',
                figure=px.pie(payment_modes, names='Payment Mode', values='Total Amount',
                              title='Payment Mode Distribution')
            )
        ]),

        dcc.Tab(label='Sales Trends', children=[
            dcc.Graph(
                id='sales-trend-chart',
                figure=px.line(sales_trend, x='Voucher Date', y='Voucher Amount', title='Daily Sales Trends')
            )
        ]),

        dcc.Tab(label='Cancellation Analysis', children=[
            dcc.Graph(
                id='cancellation-analysis-chart',
                figure=px.bar(cancellation_analysis, x='Cancel Reason', y='Voucher Amount',
                              title='Cancellation Analysis')
            )
        ])
    ])
])

# Run Dashboard
if __name__ == '__main__':
    dashboard.run_server(debug=True, port=8050)
```
